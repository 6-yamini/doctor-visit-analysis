```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("/content/Doctor-Visit-Analysis.csv")
df.head(15)
```

```python
print(df.head(15))
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df['Patient ID'].value_counts())
```

```python
plt.hist(df['Height (cm)'],bins=12, edgecolor='black')
plt.xlabel('Height(cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Patient Height with Frequency')
plt.show()
```

```python
sns.histplot(data=df, x='Age', bins=10)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Patient Age with count')
plt.show()
```

```python
sns.histplot(data=df, x='Weight (kgs)', bins=8)
plt.xlabel('Weight (kgs)')
plt.ylabel('Count')
plt.title('Distribution of Patient weight with count')
plt.show()
```

```python
sns.scatterplot(data=df, x='Visit Duration', y='Wait time (mins)')
plt.xlabel('Visit Duration')
plt.ylabel('Wait time (mins)')
plt.title('Visit Duration vs. Wait time (mins)')
plt.show()
```

```python
doctor_visits = df['Patient ID']
plt.bar(doctor_visits.index, doctor_visits.values)
plt.xlabel('Counts')
plt.ylabel('Patient ID')
plt.title('Patient ID Counts')
plt.show()
```

```python
doctor_visits = df['Appointment ID']
plt.bar(doctor_visits.index, doctor_visits.values)
plt.xlabel('Counts')
plt.ylabel('Appointment ID')
plt.title('Appointment ID Counts')
plt.show()
```

```python
sns.histplot(data=df, x='Age', y='Height (cm)', bins=5)
plt.xlabel('Age')
plt.ylabel('Height')
plt.title('Distribution of Patient Ages with Height')
plt.show()
```

```python
data= df.drop_duplicates()
print("Data:\n",data)
```

```python
average_age = np.mean(df['Age'])
print("Average Age:", average_age)
```

```python
gender_count = df['Gender'].value_counts()
print("Gender counts:\n",gender_count)
```

```python
total_duration_by_appointment_type = df.groupby('Appointment Reason')['Visit Duration'].sum()
print("Total duration by appointment:\n",total_duration_by_appointment_type)
```

```python
gender_count.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Number of Patients by Gender')
plt.show()
```

```python
df.boxplot(column='Visit Duration', by='Appointment Reason')
plt.xlabel('Appointment Reason')
plt.ylabel('Visit Duration')
plt.title('Visit Duration by Appointment Type')
plt.show()
```

```python
sns.histplot(df.Gender, bins=2)
```

```python
label=['Yes', 'No']
Y = df[df['Freepoor']=='Yes']
N = df[df['Freepoor']=='No']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x, labels=label)
plt.title("% of people getting govt health Insurance due to low income ")
plt.show()
Y = df[df['Private']=='Yes']
N = df[df['Private']=='No']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x, labels=label)
plt.title("% of people having private health Insurance ")
plt.show()
Y = df[df['Freerepat']=='Yes']
N = df[df['Freerepat']=='No']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x, labels=label)
plt.title("% of people getting govt health Insurance due to old age, disability or veteran status")
plt.show()
```

```python
db= df.groupby('Gender')['Reduced'].sum().to_frame().reset_index()
plt.barh(db['Gender'], db['Reduced'], color = ['cornflowerblue', 'lightseagreen'])
plt.title('Bar Chart')
plt.xlabel('Reduced Activity')
plt.ylabel('Gender')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
average_duration_by_date = df.groupby('Date')['Visit Duration'].mean()
plt.plot(average_duration_by_date.index, average_duration_by_date.values)
plt.xlabel('Date')
plt.ylabel('Average Visit Duration')
plt.title('Average Visit Duration over Time')
plt.show()
```

```python
appointment_type_counts = df['Appointment Reason'].value_counts()
plt.pie(appointment_type_counts.values, labels=appointment_type_counts.index,
        wedgeprops={'edgecolor': 'white'}, autopct='%1.1f%%')
circle = plt.Circle((0, 0), 0.4, color='white')
plt.gca().add_artist(circle)
plt.title('Distribution of Appointment Reason')
plt.show()
```
